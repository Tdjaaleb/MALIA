```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from datetime import date, time, datetime
```

```python
from Load_National_Data import Get_national_data
from Load_Regional_Data import Get_regional_data
from Load_Temperature import Get_temperature

from Temperature_interpolation import Complete_Temp_Date, temp_interpolation

from Covariables import covariables, fft_denoiser
```

```python
def temperature_creation(year, region):
    df = Get_temperature(year, region)
    df = Complete_Temp_Date(df, period=20209) # 20209 = nombre de périodes sur 5 ans et 11 mois + 2 jours des années bisextiles
    df = temp_interpolation(df)
    print(region, df.shape[0])
    return df
```

```python
year = ["2016", "2017", "2018", "2019", "2020", "2021", "2022"]
```

```python
temp_Aura = temperature_creation(year, region='AURA')
temp_Bourgogne = temperature_creation(year, region='Bourgogne')
temp_Bretagne = temperature_creation(year, region='Bretagne')
temp_Centre = temperature_creation(year, region='Centre')
temp_GrandEst = temperature_creation(year, region='GrandEst')
temp_HautsDeFrance = temperature_creation(year, region='HautsDeFrance')
temp_IDF = temperature_creation(year, region='IDF')
temp_Normandie = temperature_creation(year, region='Normandie')
temp_NouvelleAquitaine = temperature_creation(year, region='NouvelleAquitaine')
temp_Occitanie = temperature_creation(year, region='Occitanie')
temp_PACA = temperature_creation(year, region='PACA')
temp_PaysDeLoire = temperature_creation(year, region='PaysDeLoire')
```
